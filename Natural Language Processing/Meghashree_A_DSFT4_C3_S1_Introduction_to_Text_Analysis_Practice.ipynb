## Task 2 
- **Crawling With the Pattern Library**

```python
class Crawl(Crawler): 
    def visit(self, link, source=None):
        st = "VISITED: " + str(link.url) + " | FROM: " + str(link.referrer)
        print(st)
    def fail(self, link):
        print("failed:", repr(link.url))
        
url1 = "https://www.nature.com/"
p = Crawl(links=[url1], delay=0.2)
num_visited = 0
while (not p.done and num_visited < 10) :
    p.crawl(
        method=DEPTH, 
        cached=False, 
        throttle=1
    )
    num_visited += 1
```

## Task 3
- **Using BeautifulSoup4 to extract HTML Information**

```python
url2="https://www.nature.com/"

# Make a GET request to fetch the raw HTML content
html_content = requests.get(url2).text

# Parse the html content
soup = BeautifulSoup(html_content, "lxml")
```

```python
for link in soup.find_all("a"):
    print("Text: {}".format(link.text))
    print("Title:. {}".format(link.get("title")))
    print("href: {}".format(link.get("href")))
```
